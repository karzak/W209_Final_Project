```python
import pandas as pd
```

```python
# Get list of file names in the folder

from os import listdir
from os.path import isfile, join
usapath = 'C:/Users/jlandesman/Documents/MIDS/209 - Data Visualization/Final Project/zika/United_States/CDC_Report/data'
file_list = [f for f in listdir(usapath) if isfile(join(usapath, f)) and ".csv" in f]
usfiles = [join(usapath,file_list[i]) for i in range(len(file_list))]

print "First 5 files: "
usfiles[0:5]
```

```python
#Read in the data and Concat to a Pandas Data Frame
results = [pd.read_csv(f, parse_dates = ['report_date']) for f in usfiles] # Parse dates for each file individually!
usa = pd.concat(results) #Join into a single DF

#usa.index = usa.report_date # Set Index
usa.country = "USA"

#Look at DF:
usa.head()
```

```python
# Split Location index to states: 
indicies = pd.unique(usa['location'])
indicies= indicies[pd.notnull(indicies)]
states = [indicies[i].split('-')[1] for i in range(len(indicies))]
states[0:5]
```

```python
#Rejoin to original
merge_df = pd.DataFrame({'state': states, 'location': indicies})
usa = pd.merge(usa,merge_df, how = 'left', on = 'location')

#Look at DF:
usa.head()
```

```python
## Add in Florida

# Get list of ifles
florida_path = 'C:/Users/jlandesman/Documents/MIDS/209 - Data Visualization/Final Project/zika/United_States/FDH_Arbovirus/data'
file_list = [f for f in listdir(florida_path) if isfile(join(florida_path, f)) and ".csv" in f]
florida_files = [join(florida_path,file_list[i]) for i in range(len(file_list))]

# Read in and concatenate to a single DF
results_fl = [pd.read_csv(f, parse_dates = ['report_date']) for f in florida_files] # Parse dates for each file individually!
fl = pd.concat(results_fl) #Join into a single DF

#Rename State and country
fl['country'] = "USA"
fl['state'] = "Florida"
fl.head()

#Rejoin Fl with USA
usa = pd.concat((usa, fl))

#Look at df:
usa[usa.state == 'Florida'].head()
```

```python
#Get County data for FL: 
fl_counties = usa[usa.location_type == 'county']
indicies = pd.unique(fl_counties['location'])
indicies= indicies[pd.notnull(indicies)]
counties = [indicies[i].split('-')[2] for i in range(len(indicies))]
counties

#Rejoin to original
merge_df = pd.DataFrame({'county': counties, 'location': indicies})
usa = pd.merge(usa,merge_df, how = 'left', on = 'location')

#View counties
usa[pd.notnull(usa.county)].head()
```

```python
from sqlalchemy import *
engine = create_engine('sqlite:///zika.db') # Create the DB, output SQL commands
metadata = MetaData(bind=engine)

usa.to_sql('usa',engine, if_exists='replace')
metadata.create_all()
```
